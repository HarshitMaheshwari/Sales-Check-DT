```python
company_data['ShelveLoc'] = company_data['ShelveLoc'].map({'Bad':0,'Good':1,'Medium':2})
company_data['Urban'] = company_data['Urban'].map({'Yes':1,'No':0})
company_data['US'] = company_data['US'].map({'Yes':1,'No':0})
```

```python
company_data.head(10)
```

```python
colnames = list(company_data.columns)
```

```python
Y = colnames[0]
X = colnames[1:]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(company_data, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='gini')
```

```python
clf = model.fit(train[X],train[Y])
```

```python
pred = model.predict(test[X])
```

```python
pred
```

```python
pd.Series(pred).value_counts()
```

```python
pd.crosstab(test[Y],pred)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(clf)
plt.show()
```
